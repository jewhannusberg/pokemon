```python
import pandas

data = pandas.read_csv('Pokemon.csv')
```

```python
print data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.array([0, 1, 2, 3, 4])
x_label = list(data.columns.values[5:10])

plt.bar(x, data.ix[0][5:10].astype(float))
plt.xticks(x, x_label)
plt.show()
plt.close()
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(8, 4))
fig.tight_layout()
bulbasaur = data.ix[0][5:10]
ivysaur = data.ix[1][5:10]
venusaur = data.ix[2][5:10]
evolution = np.array([bulbasaur, ivysaur, venusaur])


for i in range(3):
    axes[i].bar(x, evolution[i].astype(float))
    axes[i].set_title(data.Name[i])
    axes[i].set(xlabel=x_label)


plt.show()
plt.close()
```
